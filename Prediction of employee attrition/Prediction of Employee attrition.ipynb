## Encoding

```python
catcol=data.select_dtypes("object").columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()

data[catcol]=oe.fit_transform(data[catcol])
```

```python
data.head()
```

## Seperate features and target

```python
y=data["left"]
```

```python
df=data.copy()
```

```python
df.drop("left",axis=1,inplace=True)
```

```python
x=df.iloc[:,:]
```

```python
x
```

```python
y
```

## seperating training and testing data

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

## model building

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy\n",accuracy_score(ytest,ypred))
print()
print("Report:\n",classification_report(ytest,ypred))
```
